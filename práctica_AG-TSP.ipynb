```python
def create(alphabet: list, N=100) -> list[list]:
    """Create and return new population"""

    population = []
    for i in range(N):
        population.append(copy.copy(alphabet))
        random.shuffle(population[i])
    return population

def sort_pop(population: list[list], fitness_function) -> tuple[list[list], list]:
    """Sort population by fitness function. Return tuple with population list and fitness list"""

    fitness_list = [fitness_function(ind) for ind in population]
    lista = sorted(zip(population, fitness_list), key=lambda x: x[1])
    return [x[0] for x in lista], [x[1] for x in lista]

def select(population: list[list], T: int) -> list:
    """Return a copy of an indivudual by tournament selection. Population already ordered by fitness"""

    pop2 = copy.copy(population[:T])
    return random.choice(pop2)

def mutate(ind: list, pmut: float) -> list:
    """Mutate an individual, swap elements. Return mutated individual"""

    if random.random() < pmut:
        ind[0], ind[5] = ind[5], ind[0]
        ind[2], ind[4] = ind[4], ind[2]
    return ind

def crossover_BACK(ind1: list, ind2: list, pcross: float) -> tuple[list, list]: 
    """Perform crossover based on PMX (Partially Mapped Crossover)"""

    if random.random() < pcross:
        start, end = 3, 7
        
        # Create copies of the parents
        child1, child2 = ind1[:], ind2[:]
        child1[start:end], child2[start:end] = ind2[start:end], ind1[start:end]
        
        for i in range(len(child1)):
            if i >= start and i < end:
                continue  # Skip swapped segment
            while child1[i] in child1[start:end]:
                # Replace duplicates with missing elements
                missing = set(range(10)) - set(child1)
                child1[i] = missing.pop()
        
        for i in range(len(child2)):
            if i >= start and i < end:
                continue  # Skip swapped segment
            while child2[i] in child2[start:end]:
                # Replace duplicates with missing elements
                missing = set(range(10)) - set(child2)
                child2[i] = missing.pop()
    else:
        child1, child2 = ind1[:], ind2[:]

    return child1, child2

def crossover(parent1: list, parent2: list, pcross: float) -> tuple[list, list]:
    """
    Performs crossover on two parents to generate two children.

    Parameters:
        parent1: The first parent, a permutation of city indices.
        parent2: The second parent, a permutation of city indices.
        pcross: The probability of performing crossover.

    Returns:
        A tuple containing two children (permutations of city indices).
    """
    if random.random() < pcross:
        start, end = sorted(random.sample(range(len(parent1)), 2))
        
        child1 = [-1] * len(parent1)
        child2 = [-1] * len(parent2)
        child1[start:end] = parent1[start:end]
        child2[start:end] = parent2[start:end]
        
        # Fill in the remaining cities, preserving order and avoiding duplicates
        def fill_child(child, parent, start, end):
            child_idx = end  # Start filling from the end of the crossover segment
            for city in parent:
                if city not in child:
                    if child_idx >= len(child):  # Wrap around if index exceeds length
                        child_idx = 0
                    child[child_idx] = city
                    child_idx += 1
        
        fill_child(child1, parent2, start, end)
        fill_child(child2, parent1, start, end) 
    else:
        child1, child2 = parent1[:], parent2[:]

    return child1, child2

def evolve(population: list[list], fitness_function, pmut: float, pcross=0.7, ngen=100, T=2, trace=0, elitism=False) -> None:
    """Genetic evolution process"""

    sorted_pop = population
    for i in range(ngen):
        sorted_pop, sorted_fitness = sort_pop(sorted_pop, fitness_function)
        selected_individual = select(sorted_pop, T)
        mutated_individual = mutate(selected_individual, pmut)
        child1, child2 = crossover(mutated_individual, random.choice(sorted_pop), pcross)

        sorted_pop[-1] = child1
        sorted_pop[-2] = child2

        if i % trace == 0:
            sorted_pop, sorted_fitness = sort_pop(sorted_pop, fitness_function)

            print(f"Num gen: {i}, Fitness óptimo: {sorted_fitness[0]}")
            # print(f"Mejor individuo: {sorted_pop[0]}")

    sorted_pop, sorted_fitness = sort_pop(sorted_pop, fitness_function)
    return sorted_pop
```

```python
# hay otras estrategias de mutación mejores para este problema, por ejemplo borrar e insertar
# la mutación consistirá en borrar un elemento e insertarlo en otro lado
# def mutate(ind: list[list[float,float]], pmut: float, L: int): 
def mutate(ind: list[list[float,float]], pmut: float): 
    if random.random() < pmut:
        L = len(ind)
        
        # Choose a random city to remove
        remove_idx = random.randint(0, L-1)
        city = ind.pop(remove_idx)
        
        # Choose a new random position to insert the city
        insert_idx = random.randint(0, L-1)
        
        # Insert the city at the new position
        ind.insert(insert_idx, city)
    
    return ind
```
