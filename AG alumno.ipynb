Haremos una implementación funcional, aunque sería más apropiado definir una clase

```python
import random
import copy

def create (alphabet, N=100): # crea y devuelve población
    ret=[]
    for _ in range(N):
        alp = copy.copy(alphabet)
        random.shuffle(alp)
        ret.append(alp)
    return ret

def sort_pop (pop : list, fit ): # devuelve una tupla: la población ordenada por fitness, y la lista de fitness
    fitness = [fit(p) for p in pop]
    b=sorted(zip(pop,fitness),key=lambda x: x[1])
    pop_sorted = [x[0] for x in b]  # Extrae la población ordenada
    fitness_sorted = [x[1] for x in b]  # Extrae la lista de fitness ordenada
    return pop_sorted, fitness_sorted



def select (pop, T): # devuelve un individuo seleccionado por torneo, devuelve una copia para evitar efectos laterales
    # pop se supone ya ordenada por fitness
    choices=random.choices(copy.copy(pop),k=T)
    indices=[pop.index(c) for c in choices]
    return pop[np.argmin(indices)]

def crossover (ind1, ind2, pcross): # devuelve el cruce (emparejamiento) de dos individuos
    child1, child2 = [],[]
    for i in range(len(ind1)):
        if random.random()<pcross:
            if ind1[i] not in child1:
                child1.append(ind1[i])
            else:
                child1.append(ind2[i])
            if ind2[i] not in child2:
                child2.append(ind2[i])
            else:
                child2.append(ind1[i])
        else:
            if ind2[i] not in child1:
                child1.append(ind2[i])
            else:
                child1.append(ind1[i])
            if ind1[i] not in child2:
                child2.append(ind1[i])
            else:
                child2.append(ind2[i])
    return child1,child2

def mutate (ind, pmut): # devuelve individuo mutado; la mutación consistirá en intercambiar elementos
    if random.random()<pmut:
        index1=random.randrange(0,len(ind))
        index2=random.randrange(0,len(ind))
        ind[index1],ind[index2]=ind[index2],ind[index1]
    return ind

def checkvalid(ind):

    tuple_subarrays = [tuple(subarray) for subarray in ind]
    
    # Compare the length of the set with the original array
    return len(tuple_subarrays) == len(set(tuple_subarrays))


def evolve (pop, fit, pmut,pcross=0.7, ngen=100, T=2, trace=0, elitism=False):
    for _ in range(ngen):
        new_pop=[]
        pop_sorted,_=sort_pop(pop,fit)
        if elitism:
            new_pop.append(pop_sorted[0])
            new_pop.append(pop_sorted[1])
            
        while len(new_pop) != 100:   
            ind1= select(pop_sorted,T)
            ind2= select(pop_sorted,T)
            new1,new2=crossover(ind1,ind2,pcross)
            if checkvalid(new1) and checkvalid(new2):
                mut1=mutate(new1,pmut)
                mut2=mutate(new2,pmut)
                new_pop.append(mut1)
                new_pop.append(mut2)
            else:
                pass
        pop=[*new_pop]
    pop_sorted,a=sort_pop(pop,fit)
    return pop_sorted
        

    
```

```python
# pero el azar siempre es nuestro aliado
# el uso de varios operadores de mutación ayuda a salir de mínimos locales

def mutate_swap (ind, pmut): # la mutación consistirá en intercambiar elementos
    if random.random()<pmut:
        index1=random.randrange(0,len(ind))
        index2=random.randrange(0,len(ind))
        ind[index1],ind[index2]=ind[index2],ind[index1]
    return ind

def mutate_insert (ind, pmut): # la mutación consistirá en borrar un elemento e insertarlo en otro lado
    if random.random()<pmut:
        index=random.randrange(len(ind))
        value=ind.pop(index)
        ind.insert(len(ind),value)
    return ind

def mutate_reverse (ind, pmut): # selecciona dos puntos e invierte la secuencia entre ellos
    if random.random()<pmut:
        idx1, idx2 = sorted(random.sample(range(len(ind)), 2))
        ind[idx1:idx2 + 1] = ind[idx1:idx2 + 1][::-1]
    return ind

mutations = [mutate_swap, mutate_insert, mutate_reverse]
def mutate (ind, pmut):
    operator = random.choice(mutations)
    return operator(ind, pmut)
```
